# Learning Pandas

### Import Pandas library

```python
import pandas as pd
import re 
```

### Read CSV file and Save as Dataframe

```python
df = pd.read_csv('pokemon_data.csv')
df.head(10)
```

### Read Headers

```python
df.columns
```

### Read each Column

```python
df[['Name', 'Type 1', 'HP']][0:10]
```

### Read each Row

```python
df.iloc[0:5]
```

### Read a Specific Location (R, C)

```python
df.iloc[10:15, 0:5]
```

### Accessing data in each row

```python
counter = 0
for index, row in df.iterrows():
    if counter < 10:
        print(index, row['Name'])
        counter += 1
    else:
        break
```

### Read Rows based on a Specific Condition

```python
df.loc[df['Type 1'] == "Grass"]
```

# Describing Data

```python
df.describe()
```

### Sorting Data

```python
df.sort_values('Name')
df.sort_values('Name', ascending=False)
df.sort_values(['Type 1', 'Type 2'])
df.sort_values(['Type 1', 'Type 2'], ascending=[1, 0])
```

### Summing Multiple Columns to a New Column

```python
# Method 1
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
# Method 2
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

df.head()
```

### Rearranging Columns

```python
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head()
```

### Saving Data (Exproting into Desired Format)

```python
df.to_csv('modified_pokemon.csv', index=False)
df.to_csv('modified_pokemon.txt', index=False, sep='\t')
df.to_excel('modified_pokemon.xlsx', index=False)
```

### Filtering Data

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
new_df.reset_index(drop = True, inplace = True)
new_df
```

```python
df.loc[df['Type 1'].str.contains('fire|grass', flags = re.I, regex = True)]
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex = True)]
```

### Conditional Changes

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
df.head(n=10)
```

```python
df.loc[df['Total'] > 500, ['Generation', 'Legendary']] = ['Test 1', 'Test 2']
df
```

### Aggregate Statistics (groupby)

```python
df.groupby(['Type 1']).mean().sort_values('Defense', ascending = False)
```

```python
df['count'] = 1
df.groupby(['Type 1']).count()['count']
```
